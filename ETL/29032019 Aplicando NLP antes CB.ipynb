### sacando las stopwords

```python
def no_stops(texto):
    
    doc = nlp(texto)
    all_stops = [w.text for w in doc if w.is_stop == False]
    return ' '.join(all_stops)
```

```python
# Apply the nouns function to the transcripts to filter only on nouns
data_no_stops = pd.DataFrame(data_no_clean.texto.apply(no_stops))
data_no_stops
```

### sacando entidades y sustantivos de una vez, aplicando chunk_nouns

esto se hace a partir del dataframe que no tiene stop_words

```python
#sacando los noun_chunks
def noun_chunks (texto):
    
    doc = nlp(texto)
    all_chunks = [w.text for w in list(doc.noun_chunks)]
    return ' '.join(all_chunks)
```

```python
# Apply the entities function to the transcripts to filter only on entities
data_noun_chunks = pd.DataFrame(data_no_stops.texto.apply(noun_chunks))
data_noun_chunks
```

```python
data_noun_chunks.get_values
```

### sacando los lemmas

lematización se aplica solo al dataframe con nouns

```python
def lemmas(texto):
    
    doc = nlp(texto)
    all_lemmas = [w.lemma_ for w in doc]
    return ' '.join(all_lemmas)
```

```python
# Apply the nouns function to the transcripts to filter only on nouns
data_lemmas = pd.DataFrame(data_noun_chunks.texto.apply(lemmas))
data_lemmas
```

### quitar los signos de puntuación
Pasar texto a minúsculas Quitar signos de puntuación (interrogación, etc.) Quitar espacios extra, cambio de carro, tabulaciones

```python
# Aplicaremos varios rounds de limpieza
import re
import string
```

```python
def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?¿¡!\]\%', ' ', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# vemos la primer limpieza
data_nlp.Nlp = pd.DataFrame(data_nlp.Nlp.apply(round1))
data_nlp
```

```python
# Segundo round
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…«»]', '', text)
    text = re.sub('\n', ' ', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
# veamos como queda
data_nlp.Nlp = pd.DataFrame(data_nlp.Nlp.apply(round2))
data_nlp
```
